```python
import numpy as np
import pandas as pd 
import matplotlib
import matplotlib.pyplot as plt
import sklearn
from sklearn.preprocessing import StandardScaler, OrdinalEncoder, LabelEncoder, OneHotEncoder, PolynomialFeatures
from sklearn.impute import SimpleImputer
from sklearn.model_selection import train_test_split, RandomizedSearchCV, cross_val_score, GridSearchCV, KFold, cross_validate
from sklearn.linear_model import Ridge, LogisticRegression, LinearRegression, Lasso, ElasticNet
from sklearn.tree import DecisionTreeClassifier, DecisionTreeRegressor
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.compose import ColumnTransformer, make_column_transformer
from sklearn.pipeline import make_pipeline, Pipeline
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.feature_selection import RFE, RFECV
from sklearn.dummy import DummyClassifier
from xgboost import XGBRegressor
import shap
import eli5 
```

```python
# car_df = pd.read_csv('vehicles.csv').sample(n=10000, replace='False')
car_df = pd.read_csv('vehicles.csv').sample(frac = 0.3)
car_df.head()
```

```python
car_df.info()
```

```python
car_df.rename(columns = {'county':'country'}, inplace = True)
```

```python
total_na = pd.DataFrame(data = {'Null Value': car_df.isna().sum(),
                                'Not_Null': len(car_df) - car_df.isna().sum(),
                                'Per_Null': (car_df.isna().sum() * 100) / len(car_df)})
total_na
```

```python
country_na = car_df.country.isna().sum()
print(f'Total Not na data in car_df are = {country_na - len(car_df)}')
```

```python
drop_columns = total_na[total_na.Per_Null > 40].index
drop_columns
```

```python
car_df.drop(columns = drop_columns, inplace = True)
car_df.drop(columns = ['id','url','region_url','VIN','image_url','description','posting_date','lat','long','state','paint_color'], inplace = True)
car_df.head()
```

```python
car_df.dropna(inplace = True)
```

```python
total_na = pd.DataFrame(data = {'Null Value': car_df.isna().sum(),
                                'Not_Null': len(car_df) - car_df.isna().sum(),
                                'Per_Null': (car_df.isna().sum() * 100) / len(car_df)})
total_na
```

```python
label_column = ['region','manufacturer','type','fuel','title_status','transmission','drive','model']
car_df[label_column] = car_df[label_column].apply(LabelEncoder().fit_transform)
car_df.head()
```

```python
X = car_df.drop(columns = 'price')
y = car_df['price']
for i in X.columns:
    plt.scatter(car_df[i], car_df['price'])
    plt.title(i)
    plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.80)
```
